```python
coffee_data = pd.read_csv('coffe_string_data.csv', sep=" ")
coffee_data
```

### 1. Rename column headers


```python
coffee_data = coffee_data.rename(columns={'ordernumber': 'Order Number',
                                          'date': 'Date',
                                          'status': 'Status',
                                            'total': 'Total and Items', 
                                            'shipping': "Shipping Method"})
coffee_data
```

### 2. Remove the "#" from the order number and convert the order number to a int64.

```python
coffee_data["Order Number"] = coffee_data["Order Number"].str.replace('#', '')
coffee_data
```

```python
coffee_data.info()
```

```python
coffee_data["Order Number"] = pd.to_numeric(coffee_data["Order Number"])
coffee_data
```

```python
coffee_data.info()
```

### 3. Convert date to datetime

```python
coffee_data["Date"] = coffee_data["Date"].astype('datetime64[ns]')
coffee_data
```

```python
coffee_data.info()
```

### 4. Convert total to two columns: "Total" column with the total as a float, "Number of Items" as an int64.

```python
total_items = coffee_data["Total and Items"].str.split(' ')
total_items
```

```python
total_amt = []
num_items = []
for item in total_items:
    total_amt.append(item[0].replace('$', ''))
    num_items.append(item[2])
```

```python
total_amt
```

```python
num_items
```

```python
coffee_data["Total"] = total_amt
coffee_data["Number of Items"] = num_items
coffee_data
```

```python
coffee_data_clean = coffee_data.drop(["Total and Items"], axis=1)
coffee_data_clean["Total"] = coffee_data_clean["Total"].astype(float)
coffee_data_clean["Number of Items"] = pd.to_numeric(coffee_data_clean["Number of Items"])
```

```python
coffee_data_clean.info()
```

```python
# Reorder the columns
coffee_data_clean = coffee_data_clean[['Order Number', 'Date', 'Status', 'Total', 'Number of Items', 'Shipping Method']]
coffee_data_clean
```
